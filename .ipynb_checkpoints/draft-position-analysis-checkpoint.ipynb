```python
import pandas as pd
import numpy as np
import math
```

```python
df = pd.DataFrame(pd.read_csv('./FantasyPros_2020_Overall_ADP_Rankings.csv'))

df.fillna(0)
```

```python
num_of_draft_spots = 12

def apply_row(row):
    if (row['Est. Round'] % 2 == 0):
        row['Pick'] = row['Rank']
    else:
        row['Pick'] = row['Rank'] / row['Est. Round'] / row['Est. Round']
    return row


df['Est. Round'] = df['Rank'].truediv(num_of_draft_spots).map(math.ceil, na_action='ignore')

df = df.apply(apply_row, axis=1)


df.head(60)
```
