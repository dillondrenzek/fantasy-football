```python
import pandas as pd
import numpy as np
import math
import re

# Panda Config
pd.set_option('display.max_rows', 1000)

qb_data = pd.read_csv('./raw-data/FantasyTiers-QBs.csv')
rb_data = pd.read_csv('./raw-data/FantasyTiers-RBs.csv')
wr_data = pd.read_csv('./raw-data/FantasyTiers-WRs.csv')
```

```python
# Remove Team from Player name
def parse_player(x):
    return re.sub(r"\ *[A-Z]{2,3}$", '', x)

qb_data['Player'] = qb_data['Player'].apply(parse_player)
rb_data['Player'] = rb_data['Player'].apply(parse_player)
wr_data['Player'] = wr_data['Player'].apply(parse_player)
```

```python
qb_df = pd.DataFrame(qb_data)
rb_df = pd.DataFrame(rb_data)
wr_df = pd.DataFrame(wr_data)

player_df = qb_df.append(rb_df, ignore_index=True).append(wr_df, ignore_index=True)

player_df.head(50)
```

```python
player_df.to_csv('./data/player-draft-tiers.csv', index=False)
```
